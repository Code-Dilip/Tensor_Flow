Housing Prices

```python
# import packages

import numpy as np
import tensorflow as tf
```

```python
#function to create train and test dataset

def create_train_dataset():

    number_of_bedrooms = np.array([2,3,1,4,2,5],dtype=float)
    prices_in_hundreds_of_thousands = np.array([3.5,4.2,2.8,5.5,3.8,6.3],dtype=float)

    return (number_of_bedrooms,prices_in_hundreds_of_thousands)
```

```python
(train_x,train_y) = create_train_dataset()
```

```python
print("train_x.shape:",train_x.shape)
print("train_y.shape:",train_y.shape)
```

```python
# function to create and compile model

def define_and_compile_model():

    model = tf.keras.Sequential([
        tf.keras.Input(shape=(1,)),
        tf.keras.layers.Dense(units=1)
    ])

    model.compile(optimizer='sgd',loss='mean_squared_error')

    return model
```

```python
model = define_and_compile_model()
print(model.summary())
```

```python
# function to train model
def train_model():

    (train_x,train_y) = create_train_dataset()
    model = define_and_compile_model()

    model.fit(train_x,train_y,epochs = 500)

    return model
```

```python
# training the model

trained_model = train_model()
```

```python
# making predictions 

print("predictions for the input 7.0 :",trained_model.predict([7.0],verbose=0))
```
